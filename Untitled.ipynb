```python
import pandas as pd
```

```python
AP = [("A", 20, 5, -2), ("B", 35, 10, 0), ("C", 55, 15, 3),("D", 65, 20, 5)]
SE = [("A", 25, 5, -1), ("B", 50, 10, 3), ("C", 70, 16, 5)]
S  = [("A", 10, 4, -3), ("B", 25, 8, 1), ("C", 40, 12, 3),("D", 65, 16, 6)]
IT  = [("A", 15, 1, -3), ("B", 25, 7, -1), ("C", 40, 13, 3),("D", 55, 19, 5)]
```

```python
vec=[]

for a in AP:
    for b in SE:
        for c in S:
            for d in IT:
                item = (a[0]+b[0]+c[0]+d[0], a[1]+b[1]+c[1]+d[1], a[2]+b[2]+c[2]+d[2],a[3]+b[3]+c[3]+d[3])
                vec.append(item)
                
                
```

```python
Lista=pd.DataFrame(vec, columns=["combinacion", "costo", "usuarios", "crecimiento"])
```
